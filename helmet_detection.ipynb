```python
import zipfile
import os

# Replace with your actual ZIP path
zip_path = '/content/drive/MyDrive/helmet_detection/Bike Helmet Detection.v1i.yolov8.zip'
extract_path = '/content/helmet_dataset'

# Unzip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Dataset extracted!")
```

```python
from ultralytics import YOLO

# Check contents to find the actual .yaml file path
!ls /content/helmet_dataset

# Example: assume 'data.yaml' is in the extracted folder
model = YOLO('yolov8n.yaml')  # you can also use yolov8s.yaml for better accuracy

# Train the model
model.train(data='/content/helmet_dataset/data.yaml', epochs=30, imgsz=640)
```

```python
model.val()
```

```python
from ultralytics import YOLO

# Load your trained model
model = YOLO('runs/detect/train/weights/best.pt')

# Run prediction on all images in a folder
model.predict(
    source='/content/helmet_dataset/test/images',  # Folder path
    save=True,
    conf=0.3,            # Confidence threshold (adjust as needed)
    imgsz=640,           # Image size
    show=False           # Set True if you want to display in notebook
)
```

```python
import os
from IPython.display import Image, display

# Path to predicted images
pred_dir = '/content/runs/detect/predict8'

# Display the first few predicted images
for img_file in os.listdir(pred_dir)[:5]:
    if img_file.endswith(('.jpg', '.png')):
        display(Image(filename=os.path.join(pred_dir, img_file)))
```
